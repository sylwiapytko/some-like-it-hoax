```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt

import matplotlib.patches as mpatches
```

```python
plt.rcParams.update({'font.size': 18})
plt.figure(figsize =(18,10)) 

typenames= ['1', '2', '>2\n<10','<10']
junk  = [1850,662,1145,560]
mainstream = [6500,1873,2718,938 ]
factcheck = [1213,300,374,107]
pies = [junk, mainstream, factcheck]
titles = ['Posty nierzetelne', 'Posty mainstream', 'Posty factcheck']
fig1, axes = plt.subplots(1,3,figsize=(14,8))

color_map =['orange', 'lightblue', 'lightpink', 'greenyellow']
for i, ax in enumerate(axes.flatten()):
    ax.pie(pies[i],  colors  = color_map, autopct="%0.1f%%",
        shadow=True, startangle=90)
    ax.set_title(titles[i])
plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=1.0)

one_patch = mpatches.Patch(color='orange', edgecolor='black', label='Jedno')
two_patch = mpatches.Patch(color='lightblue', edgecolor='black', label='Dwa')
two_ten_patch = mpatches.Patch(color='lightpink', edgecolor='black', label='Wiecej niz 2\nmniej niz 10')
ten_patch = mpatches.Patch(color='greenyellow', edgecolor='black', label='Nie mniej niz 10')
plt.legend(handles=[one_patch,   two_patch, two_ten_patch, ten_patch],
           fontsize=18,loc='upper center', bbox_to_anchor=(-0.7, +0.05),
          fancybox=True, shadow=True, ncol=5)
plt.show()
```
